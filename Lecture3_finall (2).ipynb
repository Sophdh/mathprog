Exercise4: Write a Python program to find the median of a given array of numbers with for and if loop

```python
ex4_ary = [1, 2, 3, 4, 5, 6]
ex4_len = len(ex4_ary)


if ex4_len % 2 == 0:
    median = (ex4_ary[int(ex4_len/2)] + ex4_ary[int(ex4_len/2 -1)])/2
else:
    median = ex4_ary[int(ex4_len/2)]

print(median)
        
```
