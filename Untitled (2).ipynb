# 1. Calculate the sum of all numbers from 1 to 20 using for loop

```python
start = 0
for n in range(1, 21):
       start += n
print(start)              
```

# 2. Calculate the absolute value of a negative number using if

```python
x = -4

if x < 0:
    xabs = x-2*x
    print(xabs)
else: 
    print(x)
```

# Print out the numbers from 50 to 0 in steps of 5, using while loop

```python
num = 50

while num >= 0:
    print(num)
    num -= 5
```
